```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos proporcionados
horizontes = ['1D', '1W', '1M', '3M', '6M', '1Y']
random_forests = [1.0693, 0.6050, 0.5767, 0.5298, 0.4396, 0.4140]
red_neuronal = [0.6138, 0.3433, 0.3666, 0.3086, 0.2791, 0.2540]
prophet = [0.9258, 0.5176, 0.6093, 0.5036, 0.4377, 0.4203]

# Crear un DataFrame
df = pd.DataFrame({
    'Horizontes': horizontes,
    'Random Forests': random_forests,
    'Prophet': prophet,
    'Red Neuronal': red_neuronal,
})

# Crear el gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))

# Ancho de las barras
bar_width = 0.25

# Posición de las barras en el eje X
r1 = range(len(horizontes))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Dibujar las barras
ax.bar(r1, df['Random Forests'], color='#ff7f0e', width=bar_width, edgecolor='grey', label='Random Forests')
ax.bar(r2, df['Prophet'], color='#1f77b4', width=bar_width, edgecolor='grey', label='Prophet')
ax.bar(r3, df['Red Neuronal'], color='#2ca02c', width=bar_width, edgecolor='grey', label='Red Neuronal')


# Añadir etiquetas y título
ax.set_xlabel('Horizontes', fontweight='bold')
ax.set_ylabel('MAE', fontweight='bold')
ax.set_title('Comparación de Modelos en Diferentes Horizontes Temporales')
ax.set_xticks([r + bar_width for r in range(len(horizontes))])
ax.set_xticklabels(horizontes)

# Añadir leyenda
ax.legend()

# Mostrar gráfico
plt.show()
```

```python
# Datos proporcionados
modelos = ['Random Forests', 'Prophet', 'Neural Network']
mse_values = [0.4139488994809409, 0.420325990049816, 0.25397885271730436]
mae_values = [0.46536022768315133, 0.48721402034465267, 0.3513126759171852]
mape_values = [0.5935065116966564, 0.6272790383700999, 0.4059735270880876]

# Crear el gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))

# Ancho de las barras
bar_width = 0.25

# Posición de las barras en el eje X
r1 = range(len(modelos))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Dibujar las barras con colores más suaves
ax.bar(r1, mse_values, color='#1f77b4', width=bar_width, edgecolor='grey', label='MSE')  # Azul suave
ax.bar(r2, mae_values, color='#ff7f0e', width=bar_width, edgecolor='grey', label='MAE')  # Naranja
ax.bar(r3, mape_values, color='#2ca02c', width=bar_width, edgecolor='grey', label='MAPE')  # Verde

# Añadir etiquetas y título
ax.set_xlabel('Modelos', fontweight='bold')
ax.set_ylabel('Valor de la Métrica', fontweight='bold')
ax.set_title('Comparación de Métricas para Diferentes Modelos')
ax.set_xticks([r + bar_width for r in range(len(modelos))])
ax.set_xticklabels(modelos)

# Añadir leyenda
ax.legend()

# Mostrar gráfico
plt.show()
```
